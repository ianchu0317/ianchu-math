Un globo asciende con _v_ constante de módulo 5 m/s, y 5 segundos después del despegue se dispara hacia arriba una bala de cañón con una rapidez inicial de 25 m/s. ¿Impacta la bala al globo?. **R:** No impacta en ningun instante.

```python
from sympy import *

t = Symbol('t', real=True)

# Ecuaciones del globo
v0_glob = 5 # m/s
y0_glob = 0 # m
t0_glob = 0 #s
y_glob = y0_glob + v0_glob*(t - t0_glob)

# Ecuaciones de la bala
a = -9.8 # m/s2
v0_bal = 25 # m/s
y0_bal = 0 # m
t0_bal = 5 # s
y_bal = y0_bal + v0_bal*(t - t0_bal) + (1/2)*a*(t - t0_bal)**2

# Hallar instancia en la que impactan -> y_glob(t) = y_bal(t)
inst_eq = Eq(y_glob, y_bal)
instancias = solve(inst_eq, t)
instancias
```

```python
import matplotlib.pyplot as plt
import numpy as np

y_glob = lambdify(t, y_glob)
y_bal = lambdify(t, y_bal)
t_glob = np.linspace(0, 10, 100)
t_bal = np.linspace(5, 10, 100)

plt.plot(t_glob, y_glob(t_glob), label='Globo')
plt.plot(t_bal, y_bal(t_bal), label='Bala')
plt.xlabel('Tiempo (s)')
plt.ylabel('Altura (m)')
plt.legend()
plt.grid(visible=True, linestyle=':', linewidth=0.5)
plt.show()
```
