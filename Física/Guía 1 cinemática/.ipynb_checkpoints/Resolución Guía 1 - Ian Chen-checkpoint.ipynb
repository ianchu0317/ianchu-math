## Ejercicio 2
Los móviles M y N se desplazan por trayectorias rectilíneas y paralelas. Inicialmente se encuentran en las posiciones **xM0 = 750m** y **xN0 = 0m**.

(a) Determine en qué posición y en qué instante se cruzan ambos vehículos. → *Posición 1750 en instante 50s y 3750m en instante 150s*


(b) Represente gráficamente la posición de ambos móviles en función del tiempo.





Se puede observar que N tiene un MRUV y M tiene MRU. Sabiendo eso podemos hallar las ecuaciones horarias teniendo el gráfico que nos aporta las informaciones de las velocidades y el tiempo, con el que luego buscaremos las aceleraciones de los vehículos.

```python
from sympy import *

# Variables iniciales x0
xm0 = 750
xn0 = 0

# buscar aceleraciones a = dv / dt (en m/s2)
aM = (20 - 20) / (100 - 0)
aN = (20 - 40) / (100 - 0)

# Hallar velocidades iniciales (en m/s)
vm0 = 20
vn0 = 40

# Hallar ecuaciones horarias
t  = symbols('t', real=True)
xm = xm0 + vm0*t + (1/2)*aM*(t)**2
xn = xn0 + vn0*t + (1/2)*aN*(t)**2

eq_instante = Eq(xm, xn)
instante = solve(eq_instante, t)
print('- Instante: ', instante)
print('- Posición: ', xm.subs(t,instante[1]), xm.subs(t, instante[0]))
```

```python
import matplotlib.pyplot as plt
import numpy as np

f_xm = lambdify(t, xm)
f_xn = lambdify(t, xn)
tiempo = np.linspace(0, 200, 100)

def plots(x1, x2, f1, f2, f_n1, f_n2):
    plt.plot(x1, f1(tiempo), label=f_n1)
    plt.plot(x2, f2(tiempo), label=f_n2)
    plt.grid(visible=True, linestyle=':', linewidth=0.5)
    plt.legend()
    plt.show()
    
plots(tiempo, tiempo, f_xm, f_xn, 'xM', 'xN')
```

## Ejercicio 4
Se deja caer una piedra A desde la terraza de un edificio. Cuando la piedra pasa por una
ventana ubicada a 60m debajo del nivel de la terraza, se arroja verticalmente hacia arriba
desde el piso otra piedra B, con una velocidad inicial de 20m/s. **Ambas piedras se encuentran
cuando la piedra B alcanza su altura máxima.**

(a) ¿Cuál es la altura del edificio?. R: 170.801747752989 metros

(b) Represente en un gráfico la altura de las piedras en función del tiempo. Aclare cuál es el
sistema de referencia que adopta.


```python
# Variables por enunciado
a = -9.8 # m/s2
va0 = 0 # m/s
vb0 = 20 # m/s
h = Symbol('h') # Altura h

# Ecuaciones horarias de piedra A
y_a = h + va0*(t) + (1/2)*a*t**2 

# Hallar tiempo cuando Piedra A está en H-60
t0 = solve(Eq(y_a, h-60), t)[1]

# Ecuaciones horarias de la piedra B
y_b = vb0*(t - t0) + (1/2)*a*(t - t0)**2

# Instante y altura máxima de piedra B -> donde intersecta piedra A
b_inst_max = solve(Eq(diff(y_b, t), 0), t)
b_alt_max = y_b.subs(t, b_inst_max[0])

# Hallar altura h tiendo tiempo y altura de un punto de la piedra A (donde intersecta piedra B)
altura_h = solve(Eq(y_a.subs(t, b_inst_max[0]), b_alt_max), h)
print('Altura del edificio es: ', altura_h[0])

# Gráfico
y_a = lambdify(t, y_a.subs(h, altura_h[0]))
y_b = lambdify(t, y_b)
tiempo = np.linspace(0, 6, 100)
tiempo_b = np.linspace(3.5, 7, 100)

plt.plot(tiempo, y_a(tiempo), label='Piedra A')
plt.plot(tiempo_b, y_b(tiempo_b), label='Piedra B')
plt.grid(visible=True, linestyle=':', linewidth=0.5)
plt.ylim(0, 180)
plt.legend()
plt.show()
```

## Ejercicio 6

El conductor de un automóvil que viaja a 35m/s ve, en un instante, un perro en la autopista
a 200m de  ́el. Sabiendo que tarda 0,3s en reaccionar y apretar el pedal de freno y que  ́este le
proporciona al automóvil una aceleración de frenado de 5m/s2
, calcule:

(a) El tiempo que transcurre desde que ve al perro hasta que se detiene. → Transcurre 7.3s desde que ve al perro hasta detenerse.

(b) La distancia que recorre desde que ve al perro hasta que se detiene. ¿Embiste al perro o
logra detenerse antes? → Logra detenerse a los 122.5m + 10.5m (los 0.3s antes de pisar el freno)

```python
# Datos iniciales
a = -5 # m/s2
v0 = 35 # m/s
t0 = 0.3 # s
t = Symbol('t')

# Si logra detenerse, entonces v = 0
v = v0 + a*(t - t0)
inst_frenado = solve(Eq(v, 0), t)
inst_frenado = round(inst_frenado[0], 2)
print('Tiempo transcurrido desde ver el perro hasta detenerse: ', inst_frenado)

# Distancia recorrida en su totalidad → X(inst_frenado)
x = v0*(t - t0) + (1/2)*a*(t - t0)**2
x_inicial = 35 * 0.3 # m/s
x_frenado = x.subs(t, inst_frenado)
print('Distancia recorrida desde que ve al perro hasta que se detiene: ', x_frenado + x_inicial)
```

### Ejercicio 11

Usted conduce su autom ́ovil en un camino rural con una rapidez de 27, 0m/s. Mientras sube
haca la cima de una colina observa que un tractor se encuentra en el camino a una distancia
de 35, 0m delante de su auto y movi ́endose en la misma direcci ́on y sentido que usted con una
rapidez de 10, 0m/s. Usted inmediatamente aplica los frenos y desacelera con una aceleraci ́on
constante de magnitud 7, 0m/s2.

(a) ¿Golpear ́a al tractor antes de detenerse? → **R:** No, logra detenerse antes de chocar.

(b) ¿Qu ́e tan lejos llegar ́a antes de que se detenga o choque con el tractor? → **R:** Recorrerá 52,07m antes de detenerse

(c) Si se detiene, ¿a qu ́e distancia estar ́a el tractor frente a usted en ese instante? → **R:** a 21.5m del auto

```python
# Ecuaciones horarias para el auto
a_t0 = 0 # s
a_x0 = 0 # m
a_v0 = 27 # m/s
a_a = -7 # m/s2
a_x = a_x0 + a_v0*(t - a_t0) + (1/2)*a_a*(t - a_t0)**2

# Ecuaciones horarias para el tractor
t_t0 = 0 # s
t_x0 = 35 # m
t_v0 = 10 # m/s
t_x = t_x0 + t_v0*(t - t_t0)

# X_tractor(t) = X_Auto(t) → Hallar tiempo de colision
x_eq = Eq(a_x, t_x)
t_eq = solve(x_eq, t)
print('Instantes en el que se chocan: ', t_eq)

# Hallar distancia recorrida antes de chocar Velocidad_auto(t) = 0 m/s → Hallar X_auto(t) con el tiempo despejado
a_v = a_v0 + a_a*(t - a_t0)
a_stop_t = solve(Eq(a_v, 0), t)
a_stop_t = a_stop_t[0]
print(f'Recorrerá {round(a_x.subs(t, a_stop_t), 2)} m antes de detenerse.')

# Hallar distancia entre el tractor y el auto
print(f'El tractor estará a {round(t_x.subs(t, a_stop_t) - a_x.subs(t, a_stop_t), 2)} m del auto detenido.')
```

## Ejercicio 13
Calcule el punto sin retorno para una pista de aeropuerto que tiene 2, 4km de longitud, con-
siderando que un avi ́on puede acelerar a 3m/s2 y puede desacelerar a 2m/s2

. El punto sin retorno se produce cuando el piloto ya no puede cancelar el despegue sin salirse de la pista.
¿Cu ́al es la duraci ́on del tiempo disponible para decidir libremente a partir del instante en que
el avi ́on se pone en movimiento?

```python
from sympy.physics.units import meter, second, kilometer
from sympy.physics.units import convert_to
from sympy.physics.units import Unit
import sympy.physics.units as u

x_tot = 2400 # m
t1 = symbols('t1', real=True)

# Hallar ecuaciones de aceleracion
a_acel = 3 # m/s2
v0_acel = 0 # m/s 
x0_acel = 0 # m 
t0_acel = 0 # s 
v_acel = v0_acel + a_acel*(t - t0_acel)
x_acel = x0_acel + v0_acel*(t1 - t0_acel) + (1/2)*a_acel*(t1 - t0_acel)**2

# Hallar ecuaciones de desaceleracion
a_dacel = -2 # m/s
v0_dacel = v_acel
x0_dacel = x_acel 
t0_dacel = t1
v_dacel = v0_dacel + a_dacel*(t - t0_dacel)
x_dacel = x0_dacel + v0_dacel*(t - t0_dacel) + (1/2)*a_dacel*(t - t0_dacel)**2

# Igualar velocidad de desaceleracion para que sea igual a 0 m/s (movimiento en reposo)
t_total = solve(Eq(v_dacel, 0), t)
t_total = t_total[0]
t_total

# Igualar desaceleracion = 2400m y buscar tiempo t1
t1_eq = Eq(x_dacel.subs(t, t_total), x_tot)
t1_s = solve(t1_eq, t1)
t1_s = t1_s[1] # la respuesta positiva en segundos

# Hallar punto sin retorno en metros
print('El punto sin retorno (en metros) es: ', x_acel.subs(t1, t1_s))

#Reemplazar t por tiempo total e igualarlo de nuevo a 2400 
#t1_eq = Eq(, x_tot)
# x_dacel.subs(t, t_total)
```

viaja a aproximadamente 343m/s en el aire, ¿qu ́e tan profundo es el pozo? → **R**: -46m 

```python
# Variables
h = symbols('h') # profundidad en m
t1, t2 = symbols('t1 t2') # t1 cuando piedra toca al suelo, t2 cuando tarda en subir el sonido
t_total = 3.20 # s

# Ecuacion de caida libre
t1 = t_total - t2
a_p = -9.8 # m/s2
v0_p = 0 # m/s
x0_p = 0 # m
t0_p = 0 # s
x_p = x0_p + v0_p*(t) + (1/2)*a_p*(t1 - t0_p)**2

# Ecuacion del sonido → MRU porque no avanza cada vez más rapido
v_s = 343 # m/s
x_s = h + v_s*(t2)

# Hallar tiempo en que t2 = 0 → Tiempo del subida del sonido
eq = solve(Eq(x_s.subs(h, x_p), 0), t2)
eq_t2 = eq[0]

# Hallar profundidad h reemplazando t2 en primera ecuacion
h = x_p.subs(t2, eq_t2)
h
```

## Ejercicio 16

Se lanza un cohete a escala directamente hacia arriba con una velocidad inicial de 50m/s, y
acelera a 2m/s2 de manera constante hacia arriba hasta que los motores se apagan a una altitud
de 150m.

(a) ¿Cu ́al es la altura m ́axima que alcanza el cohete? → **R**: 308.163265306122 m

(b) ¿Cu ́anto tarda el cohete despu ́es del despegue vertical en alcanzar su altura m ́axima? → **R**: 8.52021402111952 s

(c) ¿Cu ́anto tiempo permanece el cohete en el aire? → **R**: 16.4505647263136 s

```python
# Ecuaciones horarias del instante cohete con propulsión
a = 2 # m/s2
v0 = 50 # m/s
y0 = 0 # m
t0 = 0 # s
t1 = symbols('t1', real=True)
v_acel = v0 + a*(t1 - t0)
y_acel = y0 + v0*(t1 - t0) + (1/2)*a*(t1 - t0)**2

# Hallar valor de t1
t_150 = solve(Eq(y_acel, 150), t1)
t_150 = t_150[1] # la respuesta positiva

# Instante cohete sin propulsión
a = -9.8 # m/s
v0 = v_acel
y0 = 150 # m
t0 = t_150
v_dacel = v0.subs(t1, t_150) + a*(t - t0)
y_dacel = y0 + v0.subs(t1, t_150)*(t - t0) + (1/2)*a*(t - t0)**2

# Hallar tiempo total cuando velocidad = 0
t_alt_max = solve(Eq(v_dacel, 0), t)[0]
print('Tiempo para alcanzar altura máxima: ', t_alt_max, 's')

# Hallar altura máxima teniendo el tiempo
print('Altura máxima alcanzada: ', y_dacel.subs(t, t_alt_max), 'm')

# Hallar tiempo total, es decir valor de t cuando y_acel = 0
t_total = solve(Eq(y_dacel, 0), t)
t_total = t_total[1]
print('Tiempo total en el aire: ', t_total, 's')
```

## Ejercicio 17
Un astronauta con su traje puesto salta en la Tierra con una cierta velocidad inicial y logra
despegarse del suelo hasta una distancia de 20cm. Dicho astronauta viaja a Marte y efect ́ua
un salto con la misma velocidad inicial. La altura alcanzada en este caso es de 0, 525m.

(a) ¿Cu ́al es la velocidad inicial del salto?

(b) ¿Cu ́al es el valor de la aceleraci ́on de la gravedad en Marte?

```python
# Salto en tierra
a = -9.8 # m/s2
v0 = Symbol('v0')
y0 = 0 # m
y_max = 0.2 # m
v = v0 + a*(t)
y = y0 + v0*(t) + (1/2)*a*(t)**2

# Hallar tiempo de altura maxima
t_alt_max = solve(Eq(y, y_max), t)[1]

# Hallar velocidad inicial reemplazando el tiempo por el tiempo de altura máxima en la ecuacion de velocidad e igualandolo a 0 m/s
eq_v = Eq(v.subs(t, t_alt_max), 0)
v0 = solve(eq_v, v0)[1]
print('Velocidad inicial de salto en tierra es: ', v0)

# Hallar aceleracion de gravedad de marte → velocidad inicial de salto es aproximadamente el mismo que en tierra, lo que cambia es la gravedad
a = Symbol('a')
y_max = 0.525 # m
v = v0 + a*t
y = y0 + v0*t + (1/2)*a*t**2

# Haciendo el mismo procedimiento → Hallar tiempo de altura maxima
t_alt_max = solve(Eq(y, y_max), t)[0]

# Hallar aceleracion reemplazando en tiempo por el tiempo de altura maxima en la ecuacion de velocidad e igualando a 0 m/s
eq_a = Eq(v.subs(t, t_alt_max), 0)
a = solve(eq_a, a)[0]
print('Valor de aceleración en marte es: ', a)
```
