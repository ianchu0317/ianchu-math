## Ejercicio 2
Los móviles M y N se desplazan por trayectorias rectilíneas y paralelas. Inicialmente se encuentran en las posiciones **xM0 = 750m** y **xN0 = 0m**.

(a) Determine en qué posición y en qué instante se cruzan ambos vehículos. → *Posición 1750 en instante 50s y 3750m en instante 150s*


(b) Represente gráficamente la posición de ambos móviles en función del tiempo.





Se puede observar que N tiene un MRUV y M tiene MRU. Sabiendo eso podemos hallar las ecuaciones horarias teniendo el gráfico que nos aporta las informaciones de las velocidades y el tiempo, con el que luego buscaremos las aceleraciones de los vehículos.

```python
from sympy import *

# Variables iniciales x0
xm0 = 750
xn0 = 0

# buscar aceleraciones a = dv / dt (en m/s2)
aM = (20 - 20) / (100 - 0)
aN = (20 - 40) / (100 - 0)

# Hallar velocidades iniciales (en m/s)
vm0 = 20
vn0 = 40

# Hallar ecuaciones horarias
t  = symbols('t')
xm = xm0 + vm0*t + (1/2)*aM*(t)**2
xn = xn0 + vn0*t + (1/2)*aN*(t)**2

eq_instante = Eq(xm, xn)
instante = solve(eq_instante, t)
print('- Instante: ', instante)
print('- Posición: ', xm.subs(t,instante[1]), xm.subs(t, instante[0]))
```

```python
import matplotlib.pyplot as plt
import numpy as np

f_xm = lambdify(t, xm)
f_xn = lambdify(t, xn)
tiempo = np.linspace(0, 200, 100)

plt.plot(tiempo, f_xm(tiempo), color='red', label='xM')
plt.plot(tiempo, f_xn(tiempo), color='blue', label='xN')
plt.grid(visible=True, linestyle=':', linewidth=0.5)
plt.legend()
plt.show()
```

## Ejercicio 4
Se deja caer una piedra A desde la terraza de un edificio. Cuando la piedra pasa por una
ventana ubicada a 60m debajo del nivel de la terraza, se arroja verticalmente hacia arriba
desde el piso otra piedra B, con una velocidad inicial de 20m/s. **Ambas piedras se encuentran
cuando la piedra B alcanza su altura máxima.**

(a) ¿Cuál es la altura del edificio?

(b) Represente en un gráfico la altura de las piedras en función del tiempo. Aclare cuál es el
sistema de referencia que adopta.


```python
vb0 = 20 # m/s
vba = 0 # m/s → porque cae de un edificio con velocidad inicial 0?
a = -9.8 # m/s2

# Hallar altura máxima de la piedra b
y_piedra_b = 0 + vb0*(t) + (1/2)*(a)*(t)**2
v_piedra_b = diff(y_piedra_b, t)
b_inst_max = solve(Eq(v_piedra_b, 0), t)
b_y_max = y_piedra_b.subs(t, round(b_inst_max[0], 2))
b_y_max


# Hallar instante 
h, t0 = symbols('h t0')
y_piedra_a = h + vba*(t - t0) + (1/2)*a*(t - t0)**2
alt_inicial = h - 60 # Cuando t=0s
t_inicial = solve(Eq(y_piedra_a.subs(t, 0), alt_inicial), t0)
t_inicial[1]

# Hallar h teniendo en cuenta 
y_piedra_a = h + vba*(t - t_inicial[1]) + (1/2)*a*(t - t_inicial[1])**2
altura_edificio = solve(Eq(y_piedra_a.subs(t, b_inst_max[0]), b_y_max), h)
```
