Trace el gráfico de las siguientes funciones cuadráticas:

```python
import numpy as np
import matplotlib.pyplot as plt

# establecer funciones matematicas para obtener y
f1 = lambda x: x**2
f2 = lambda x: -2*x**2
f3 = lambda x: x**2-3
f4 = lambda x: -(x - 5)**2 

# funcion para graficar 
def _plot(ax, x_data, y_data, f_name):
    ax.plot(x_data, y_data)
    ax.set_xlabel('Eje X', fontsize=8)
    ax.set_ylabel('Eje Y', fontsize=8)
    ax.set_title(f_name, fontsize=10)
    ax.grid(visible=True, linestyle=':', linewidth=0.5)
    return ax

eje_x = np.linspace(0, 10, 100)
fig, axes = plt.subplots(2, 2, figsize=(10, 5))
ax1, ax2, ax3, ax4 = axes.flatten() # unpack subplots
fig.tight_layout(pad=2.5) # establecer espacios entre graficos

_plot(ax1, eje_x, f1(x), "Funcion a")
_plot(ax2, eje_x, f2(x), "Funcion b")
_plot(ax3, eje_x, f3(x), "Funcion c")
_plot(ax4, eje_x, f4(x), "Funcion d")

plt.show()
```
