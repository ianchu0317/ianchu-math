### Ejercicio 11
Trace el gráfico de las siguientes funciones cuadráticas:

```python
import numpy as np
import matplotlib.pyplot as plt

# establecer funciones matematicas para obtener y
f1 = lambda x: x**2
f2 = lambda x: -2*x**2
f3 = lambda x: x**2-3
f4 = lambda x: -(x - 5)**2 

# funcion para graficar 
def _plot(ax, x_data, y_data, f_name):
    ax.plot(x_data, y_data)
    ax.set_xlabel('Eje X', fontsize=8)
    ax.set_ylabel('Eje Y', fontsize=8)
    ax.set_title(f_name, fontsize=10)
    ax.grid(visible=True, linestyle=':', linewidth=0.5)
    return ax

eje_x = np.linspace(0, 10, 100)
fig, axes = plt.subplots(2, 2, figsize=(10, 5))
ax1, ax2, ax3, ax4 = axes.flatten() # unpack subplots
fig.tight_layout(pad=2.5) # establecer espacios entre graficos

_plot(ax1, eje_x, f1(eje_x), "Funcion a")
_plot(ax2, eje_x, f2(eje_x), "Funcion b")
_plot(ax3, eje_x, f3(eje_x), "Funcion c")
_plot(ax4, eje_x, f4(eje_x), "Funcion d")

plt.show()
```

### Ejercicio 12
12: Para las siguientes funciones cuadráticas determine en qué intervalo crece, en qué intervalo decrece, dónde es positiva, dónde es negativa, en qué puntos se anula y en qué punto alcanza su extremo:

```python
# https://stackoverflow.com/questions/37100053/how-to-define-a-mathematical-function-in-sympy
# https://www.geeksforgeeks.org/python-sympy-lambda-method/
# Inspiracion: https://stackoverflow.com/questions/39192643/finding-the-minimum-of-a-function-on-a-closed-interval-with-python
# inecuaciones: https://stackoverflow.com/questions/76335403/solve-inequality-in-sympy
# inecuaciones sympy: https://docs.sympy.org/latest/guides/solving/reduce-inequalities-algebraically.html#solving-guide-inequalities
# Lambdify: https://www.tutorialspoint.com/sympy/sympy_lambdify_function.htm
# Sympy intervalos: https://docs.sympy.org/latest/modules/sets.html#sympy.sets.sets.Union


import sympy as sp

sp.init_printing()

# Definir funciones simbólicas con sympy para análisis
x = sp.Symbol('x')

f1 = -2*x**2
f2 = -2*x*(x - 3)
f3 = -2*x**2 + x
f4 = x**2 + 2 + x + 1
f5 = -2*(x + 3)*(x - 5)

funciones = [f1, f2, f3, f4, f5]

# Clase para analizar funcion
class AFuncion:
    
    def __init__(self, f):
        self.f = f # funcion
        self.f_ = sp.lambdify(x, self.f)
        self.f_prima = self.f.diff(x)
        self.f_prima_ = sp.lambdify(x, self.f_prima) # derivada de f
        self.ceros = None
        self.ceros_prima = None
        self.int_creciente = None
        self.int_decreciente = None
        self.maxs = []
        self.mins = []
        self.pos = sp.reduce_inequalities(self.f > 0, x)
        self.neg = sp.reduce_inequalities(self.f < 0, x)
        
    # hallar ceros de una funcion
    def f_ceros(self, f):
        eq = sp.Eq(f, 0)
        ceros = sp.solve(eq, x)
        return ceros

    # Funcion para hallar intervalos crecientes / decrecientes por medio de la positividad de la pendiente
    '''
Planeamiento :
tiene un contador para cada posicion del array de ceros (DE LA FUNCION DERIVADA) . 
el algoritmo va de primero al ultimo registrando la pendiente del punto un valor antes del max / min (donde pendiente es 0).
cuando ya llega al ultimo tiene que registrar para la pendiente del punto de un valor despues del max / min (donde pendiente es 0).
Finalmente, si pendiente es positiva el int creciente se añade el intervalo analizado. 
Por ultimo se resta todos los reales (int decreciente) al int creciente. Asumiendo que todas las funciones son CONTINUAS y va de R->R
'''
    def f_int(self):      
        pendiente = None
        creciente = sp.S.EmptySet
        decreciente = sp.Interval(-sp.oo, sp.oo)
        contador = 0

        while contador <= len(self.ceros_prima):
            if contador == len(self.ceros_prima):
                pendiente = self.f_prima_(self.ceros_prima[contador - 1] + 1)
            else:
                pendiente = self.f_prima_(self.ceros_prima[contador] - 1)
    
            if pendiente > 0: 
                if contador == 0:
                    creciente = sp.Union(creciente, sp.Interval.open(-sp.oo, self.ceros_prima[contador]))
                elif contador == len(self.ceros_prima):
                    creciente = sp.Union(creciente, sp.Interval.open(self.ceros_prima[contador - 1], sp.oo))
                else:
                    creciente = sp.Union(sp.Interval.open(self.cero_prima[contador - 1], self.ceros_prima[contador]))
                
            contador += 1
        self.int_creciente = creciente
        self.int_decreciente = decreciente - creciente

    # Hallar máximo y minimo de la funcion (indicar punto en tupla)
    def f_max_min(self):
        for valor_x in self.ceros_prima:
            if self.f_prima_(valor_x - 1) > 0 and self.f_prima_(valor_x + 1) < 0:
                self.maxs.append((valor_x, self.f_(valor_x)))
            else:
                self.mins.append((valor_x, self.f_(valor_x)))
    

    def data(self):
        self.ceros = self.f_ceros(self.f)
        self.ceros_prima = self.f_ceros(self.f_prima)
        self.f_int()
        self.f_max_min()
        
        print('- Función: ', self.f)
        print('- Función derivada: ', self.f_prima)
        print('- Ceros de la función: ', self.ceros)
        print('- Valor x de Máximos / mínimos: ', self.ceros_prima)
        print('- Máximos: ', self.maxs)
        print('- Mínimos: ', self.mins)
        print('- Intervalo creciente: ', self.int_creciente)
        print('- Intervalo decreciente: ', self.int_decreciente)
        print('- Conjunto positividad: ', self.pos)
        print('- Conjunto negatividad: ', self.neg)
        

for funcion in funciones:
    print('\n----------------------------\n')
    analisis = AFuncion(funcion).data()
    
```

### Ejercicio 13
EJERCICIO 14: Represente gráficamente las siguientes funciones.


Analice en cada caso, la monotonía.

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *


# Definir las funciones
x = Symbol('x')

f1 = x**3
f2 = (x - 2)**3
f3 = x**3 - 1
f4 = x**4

f1_ = lambdify(x, f1)
f2_ = lambdify(x, f2)
f3_ = lambdify(x, f3)
f4_ = lambdify(x, f4)

x_value = np.linspace(0, 10, 1000)


# Funcion para graficar
def fplot(ax, f_name, f):
    ax.plot(x_value, f(x_value), color='green')
    ax.set_xlabel('x')
    ax.set_ylabel('y = f(x)')
    ax.set_title(f'f(x)={f_name}')


# Configurar gráficos
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 8))

fig.tight_layout(pad=4) # establecer espacios entre graficos

fplot(ax1, f1, f1_)
fplot(ax2, f2, f2_)
fplot(ax3, f3, f3_)
fplot(ax4, f4, f4_)

plt.show()
```
