### Ejercicio 1

<div> Represente en la recta numérica:</div>


```python
# https://numpy.org/doc/stable/reference/generated/numpy.array.html
# https://numpy.org/doc/stable/reference/generated/numpy.zeros_like.html
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt, pi

# Establecer array
a_num = np.array([-5, -1, 3, 6, 3/8, 1 + 2/5, 1 - 2/5, -sqrt(2), 
         sqrt(2) + 1, sqrt(2) - 1, -sqrt(2) + 1, -sqrt(2) - 1])
b_num = np.array([-3, -2, -1, 0, 1, 2, 3, -pi, -pi/2, pi/2, pi, 3*pi/2, 3.14, -3.14])

y_a_num = np.zeros_like(a_num)
y_b_num = np.zeros_like(b_num)

# Graficar recta
def graph(ax, r_name, x_num, y_num, xlim):
    ax.plot(x_num, y_num, "o", color='red')
    ax.set_title(r_name)
    ax.set_xlabel("Numeros reales")
    ax.set_xlim((-xlim, xlim))
    ax.set_ylim((-1, 1))
    
    count = 1 # para flechas alternadas porque se colisionan los numeros y no se lee
    
    for x_punto in np.sort(x_num):
        if count % 2 == 0:
            ax.annotate(f'{round(x_punto, 2)}', (x_punto, 0), xytext=(x_punto, -0.25), arrowprops=dict(width=2.5, shrink=3))
        else:
            ax.annotate(f'{round(x_punto, 2)}', (x_punto, 0), xytext=(x_punto, 0.25), arrowprops=dict(width=2.5, shrink=3))
        count += 1
    
    ax.grid(visible=True, linewidth=0.5, linestyle=':')


fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 10))
fig.tight_layout(pad=4)
graph(ax1, "Recta a", a_num, y_a_num, 6)
graph(ax2, "Recta b", b_num, y_b_num, 4)
plt.show()
```

### Ejercicio 3

 - Represente en la recta los siguientes conjuntos.
 - Limitaciones del programa es que no puede graficar hasta el infinito.
 - Hacer solo los ejercicios que se pueden hacer.

```python
import sympy as sp

a_int = np.array([2, 4])
b_int = np.array([3, 6]) 

# Funcion de graficar los intersecciones de intervalos -> verlo graficamente
def f_plot_int(title, a_int, b_int, xlim):
    fig, ax = plt.subplots(figsize=(20, 10))
    ax.set_title(title)
    
    ax.plot(a_int, [x + 0.025 for x in np.zeros_like(a_int)], linewidth=3, linestyle=':', color='red')
    ax.plot(b_int, np.zeros_like(b_int), linewidth=3, linestyle=':', color='cyan')
    
    ax.set_xlabel('Numeros reales')
    ax.set_xlim((xlim[0], xlim[1]))
    ax.set_ylim((-1, 1))

    len_a = len(a_int) - 1
    len_b = len(b_int) - 1
    ax.annotate(f'{a_int[0]}', (a_int[0], 0), xytext=(a_int[0], -0.25), arrowprops=dict(width=2.5, shrink=3))
    ax.annotate(f'{a_int[len_a]}', (a_int[len_a], 0), xytext=(a_int[len_a], -0.25), arrowprops=dict(width=2.5, shrink=3))
    ax.annotate(f'{b_int[0]}', (b_int[0], 0), xytext=(b_int[0], -0.25), arrowprops=dict(width=2.5, shrink=3))
    ax.annotate(f'{b_int[len_b]}', (b_int[len_b], 0), xytext=(b_int[len_b], -0.25), arrowprops=dict(width=2.5, shrink=3))
    
    ax.grid(visible=True, linestyle=':', linewidth=0.5)
    
    plt.show()

# Graficar para uniones de intervalos -> pasar solo el resultado
def f_plot_union(title, a_int, xlim):
    fig, ax = plt.subplots(figsize=(20, 10))
    ax.set_title(title)
    
    ax.plot(a_int, np.zeros_like(a_int), linestyle=':', linewidth=3, color='purple')
    
    ax.set_xlabel('Numeros reales')
    ax.set_xlim((xlim[0], xlim[1]))
    ax.set_ylim((-1, 1))

    len_a = len(a_int) - 1
    ax.annotate(f'{a_int[0]}', (a_int[0], 0), xytext=(a_int[0], -0.25), arrowprops=dict(width=2.5, shrink=3))
    ax.annotate(f'{a_int[len_a]}', (a_int[len_a], 0), xytext=(a_int[len_a], -0.25), arrowprops=dict(width=2.5, shrink=3))
    
    ax.grid(visible=True, linestyle=':', linewidth=0.5)
    
    plt.show()
    

f_plot_int('Ejercicio 3a', a_int, b_int, (1, 7))
```

```python
a_int = np.linspace(2, 6, 100)
f_plot_union("Ejercicio 3b", a_int, (1, 7))
```

```python
a_int = np.linspace(-1, 5, 100)  # (-1, 3) U (3, 5)
f_plot_union('Ejercicio 3f', a_int, (-2, 6))
```
