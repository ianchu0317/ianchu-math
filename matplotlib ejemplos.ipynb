```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib as mpl
```

```python
x = [1, 2, 3, 4] #coordenadas para eje x
y = [1, 4, 2, 3] # coordenadas para eje y

fig, ax = plt.subplots() # crear 
ax.plot(x, y)
plt.show()
```

```python
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
plt.show()
```

```python
x1 = np.linspace(0, 10, 100)
y1 = x1**3
y2 = 3*x1
plt.figure()
plt.title("Prueba de matplotlib")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.plot(x1, y1, linewidth=2, label="x1**3", color='yellow', linestyle=':')
plt.plot(x1, y2, linewidth=2, label="3*x", color='blue')
plt.legend()
plt.grid(True)
plt.show()
```
